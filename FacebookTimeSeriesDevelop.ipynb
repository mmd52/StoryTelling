### Build the plot

First we need to create a `Plot` object. We'll start with a basic frame, only specifying things like plot height, width, and ranges for the axes.

```python
#p = figure(x_axis_type="datetime", x_axis_label='Date' , y_axis_label='Stock price ($)' ,plot_width=800, plot_height=350)
#p.line('Date', 'Close', source=source)

xdr = Range1d(2010, 2019)
ydr = Range1d(100, 200)
plot = figure(
    #x_range=xdr,
    #y_range=ydr,
    plot_width=900,
    plot_height=500,
    outline_line_color=None,
    toolbar_location=None, 
    min_border=20,x_axis_type="datetime",
    x_axis_label='Date' ,
    y_axis_label='Stock price ($)'
)
```

```python
show(plot)
```

```python
plot.line('Date', 'Close', source=source)
show(plot)
```

In the above, `plot.add_glyph` returns the renderer, which we can then pass to the `HoverTool` so that hover only happens for the bubbles on the page and not other glyph elements:

```python
from bokeh.models import HoverTool 
hover = HoverTool(tooltips=[('Price','@Close')])
plot.add_tools(hover)
```
